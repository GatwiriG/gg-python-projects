```python
#1.Use spark to load the dataset in python and display the first fifteen transactions.
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('Kenya Ecommerce Transactions Analysis').getOrCreate()
```

```python
dataset=spark.read.csv('/kenya_ecommerce_transactions.csv',header=True,inferSchema=True)
```

```python
dataset.show(15)
```

```python
#2.Outline the data types for all the variables into either numerical or categorical (justify using code)
dataset.printSchema()
#Numericals;price,quantity,transaction_date,user_age
#Categoricals;transaction,user and product ids,category,payment method,country,city,device type,gender,1st purchase
```

```python
#3.Compute the variance of the clients' ages
from pyspark.sql.functions import variance
dataset.select(variance('user_age')).show()
```

```python
#4.What was the average age of the clients?
from pyspark.sql.functions import avg
dataset.select(avg('user_age')).show()
#average=mean=41.012
```

```python
#5.Obtain the range in prices of commodities.
dataset.describe('price').show()
#range=max-min=4999.89-100.02=4899.87
```

```python
#6.How many transactions were completed successfully?
dataset.describe('transaction_id').show()
#=100000
```

```python
#7.Obtain the clients' range in age.
dataset.describe('user_age').show()
#64-18
```

```python
#8.How many transactions were made using bank transfer?
dataset.groupBy('payment_method').count().show()
#25043
```

```python
#9.Retrieve all transactions where price was more than 5000.
dataset.filter(dataset.price>5000).count()
```

```python
#10.Retrieve transactions made using Mobile devices.
dataset.groupBy('device_type').count().show()
#33343
```

```python
#11.How many first-time purchases are there?
dataset.groupBy('is_first_purchase').count().show()
#49961
```

```python
#12.Retrieve transactions from the city of Eldoret.
dataset.groupBy('city').count().show()
#10006
```

```python
#13.Retrieve transactions where quantity was less than two.
dataset.filter(dataset.quantity<2).count()
#24842
```
