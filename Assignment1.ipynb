iii. Check for missing values and propose handling strategy

```python
dataset.isnull().sum()
```

```python
# For numeric columns I use fill with mean
numeric_cols = ['Age', 'BMI', 'Hospital_Visits_Per_Year', 'Exercise_Frequency']
for col in numeric_cols:
    dataset[col] = dataset[col].fillna(dataset[col].mean())

# For categorical columns I use fill with mode
categorical_cols = ['Gender', 'Blood_Pressure', 'Cholesterol_Level', 'Diabetes', 'Medication_Adherence', 'Smoking_Status']
for col in categorical_cols:
    dataset[col] = dataset[col].fillna(dataset[col].mode()[0])

```

```python
dataset.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Pairplot for Age, BMI, Exercise_Frequency
sns.pairplot(dataset, vars=['Age', 'BMI', 'Exercise_Frequency'], hue='Gender',diag_kind='kde')
plt.show()
```

v. Visualize relationship between BMI and Diabetes

```python
sns.stripplot(x='Diabetes',y='BMI',hue='Gender',data=dataset)
plt.title('BMIS VS DIABETES STATUS BY GENDER')
```

```python
vi. Visualize Exercise_Frequency vs Hospital_Visits_Per_Year
```
