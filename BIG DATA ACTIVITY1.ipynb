```python
dataset=pd.read_csv(r"C:\Users\USER\Downloads\student_performance.csv")
```

```python
sns.heatmap(dataset.isnull())
plt.figure(figsize=(8,8))
#the best-shows exactly where nulls are
```

```python
plt.figure(figsize=(5,5))
plt.title("missing values per variables")
plt.xticks(rotation=80)
sns.barplot(x=missing_values.index,y=missing_values.values)
sns.set_style('whitegrid')
plt.xlabel('columns')
plt.ylabel('values')
plt.show()
```

-Propose the different strategies for handling missing values and justify your preferred approach and apply your chosen strategy in Python.

```python
#categorical-mode
dataset['Gender']=dataset['Gender'].fillna(dataset['Gender'].mode()[0])
dataset['Program']=dataset['Program'].fillna(dataset['Program'].mode()[0])

```

```python
#numerical-mean
dataset['Age']=dataset['Age'].fillna(dataset['Age'].mean())
dataset['Study_Hours_Per_Week']=dataset['Study_Hours_Per_Week'].fillna(dataset['Study_Hours_Per_Week'].mean())
dataset['Attendance_Percentage']=dataset['Attendance_Percentage'].fillna(dataset['Attendance_Percentage'].mean())
dataset['CAT_Score']=dataset['CAT_Score'].fillna(dataset['CAT_Score'].mean())
dataset['Final_Exam_Score']=dataset['Final_Exam_Score'].fillna(dataset['Final_Exam_Score'].mean())
```
