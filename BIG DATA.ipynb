dropping nulls:1.DROP

                4.MEASURE OF CENTRAL TENDENCY(MODE,MEAN,MEDIAN)

```python
dataset.head()
```

```python
#forwardfill we shld have 40.3
#dataset.culmen_length_mm.fillna(method= 'ffill')-got error coz new panda vesrion doesnt allow method
dataset['culmen_length_mm'] = dataset['culmen_length_mm'].ffill()
```

```python
dataset.head()
```

```python
#backwardfill
dataset['culmen_depth_mm'] = dataset['culmen_depth_mm'].bfill()
```

```python
dataset.flipper_length_mm.mean()
```

```python
#library for statistical analysis
import scipy
from scipy import stats as st
```

pearson came with a formula that;
correlation coefficient(r):between 0.7=0.9=+ve,0.4-0.6=moderate.0.1-0.3=weak.same for negative
1=perfect +ve correlation=1=perfect -ve
0=no correlation at all


```python
#since our plot aint giving exact values we use pearson
#note,if you have missing values you will get error.use dataset.dropna(inplace=True)
st.pearsonr(dataset.culmen_length_mm,dataset.body_mass_g)
#0.59=moderate +ve correlation btn the 2.means 58% of the datapoint are correlated.
#the remaining ones are the ones causing the points look scattered coz they have no relationship
#so,at 1st with the graph we cld jump into a conclusion that they dont have any relationship but,
#after this we see that it is there but moderate.
```
